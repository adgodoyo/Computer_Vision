```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Crear la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Configurar el rango para el gráfico
z = np.linspace(1, 10, 100)  # Profundidad

# Líneas paralelas en z para x = -1 y x = -2
y = np.zeros_like(z)  # Altura constante para ambas líneas
x1 = -1 * np.ones_like(z)  # Línea en x = -1 paralela a z
x2 = 1 * np.ones_like(z)  # Línea en x = -2 paralela a z

# Dibujar las líneas paralelas
ax.plot(x1, y, z, label="Línea paralela x=-1", color="blue")
ax.plot(x2, y, z, label="Línea paralela x=1", color="green")

# Proyección hacia el punto de fuga (modelo pinhole)
z_projected = z
x1_projected = -1 / z_projected  # Línea proyectada para x = -1
x2_projected = 1 / z_projected  # Línea proyectada para x = -2

# Dibujar las líneas proyectadas
ax.plot(x1_projected, y, z_projected, label="Proyección x=-1", linestyle="--", color="red")
ax.plot(x2_projected, y, z_projected, label="Proyección x=-2", linestyle="--", color="orange")

# Etiquetas y configuración del gráfico
ax.set_xlabel("X (Ancho)")
ax.set_ylabel("Y (Altura)")
ax.set_zlabel("Z (Profundidad)")
ax.set_title("Modelo Pinhole en 3D")
ax.legend()

# Ajustar la vista para ver el gráfico "de frente"
ax.view_init(elev=0, azim=90)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
z_projected = np.linspace(1, 10, 100)  # Profundidad proyectada
x1 = 1 * np.ones_like(z_projected)  # Línea x = 1
x2 = -1 * np.ones_like(z_projected)  # Línea x = -1
y1 = 1  # Altura constante en y = 1

# Proyecciones en el plano cartesiano
x1_projected = x1 / z_projected  # Línea proyectada para x = 1
x2_projected = x2 / z_projected  # Línea proyectada para x = -1
y_projected = -y1 / z_projected  # Altura constante en y = 1 para proyección

# Configuración del gráfico en el plano cartesiano
plt.figure(figsize=(8, 6))

# Graficar las proyecciones
plt.plot(x1_projected, y_projected, label="Proyección x = 1", color='orange', linestyle='--')
plt.plot(x2_projected, y_projected, label="Proyección x = -1", color='red', linestyle='--')

# Etiquetas y configuración
plt.xlabel("X (Proyección)")
plt.ylabel("Y (Altura)")
plt.title("Proyección de Líneas en el Plano Cartesiano")
plt.axhline(0, color='black', linewidth=0.5, linestyle='dotted')  # Eje x
plt.axvline(0, color='black', linewidth=0.5, linestyle='dotted')  # Eje y
plt.grid(alpha=0.3)
plt.legend()

# Mostrar el gráfico
plt.show()


```
